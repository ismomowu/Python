```python
import numpy as np
import pandas as pd

import urllib.request
country_url = "http://rate.bot.com.tw/xrt?Lang=zh-TW"
f = urllib.request.urlopen(country_url)
html_doc = f.read()

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')    
rating_css = '.print_show' 
rating_list = soup.select(rating_css)  
    
countrylist=[]
for s in rating_list:  
      countrylist.append(s.text)

countrylist
```

```python
code=[]
for a in range(19):
    b=countrylist[a][-34:-31]
    code.append(b)
code
```

```python
import pandas
exchagerate=[]

for country in code:
    url=pandas.read_html('http://rate.bot.com.tw/xrt/quote/l6m/'+ country)
    ratetable=url[0]
    ratetable.head()
    currency=ratetable.iloc[:,0:6]
    currency.columns=["掛牌日期",'幣別',"現金賣給銀行","現金買匯","即期賣給銀行","即期買匯"]
    exchagerate.append(currency)
```

```python
exchagerate[0]
```

```python
import pandas
horizontalrate= pd.DataFrame()  

for country in code:
    url=pandas.read_html('http://rate.bot.com.tw/xrt/quote/l6m/'+ country)
    ratetable=url[0]
    ratetable.head()
    curr=ratetable.iloc[:,0:4]
    curr.columns=["掛牌日期",'幣別',country+"現金賣給銀行",country+"現金買匯"]
    horizontalrate=pd.concat([horizontalrate,curr],axis=1)
```

```python
raw= pd.DataFrame()

for country in code:
    cash=country+"現金賣給銀行"
    spot=country+"現金買匯"
    hor=horizontalrate.loc[:,[cash,spot]]
    raw=pd.concat([raw,hor],axis=1)
raw
```

```python
Date=curr.iloc[:,[0]]
countryrate=pd.concat([Date,raw],axis=1)
countryrate
```

```python
USD=raw.iloc[:128,[1]].reset_index(drop=True)
Others=raw.iloc[1:,2:].reset_index(drop=True)
predi=pd.concat([USD,Others],axis=1)
predi.pct_change().corr(method='pearson')
```

```python
X=predi.loc[:,['VND現金買匯','HKD現金買匯']]
y=predi.loc[:,['USD現金買匯']]

  
from sklearn.cross_validation import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=5)  
print(X_train.shape)
print(y_train.shape)  
print(X_test.shape)  
print(y_test.shape)  
```

```python
from sklearn.linear_model import LinearRegression  

linreg = LinearRegression()  
model=linreg.fit(X_train, y_train)  
print(model)
print(linreg.intercept_) 
print(linreg.coef_)
```

```python
feature_cols =list(X)
list(zip(feature_cols, linreg.coef_))
```

```python
y_pred = linreg.predict(X_test)  
print(y_pred)
print(type(y_pred))
```

```python
print(len(y_pred),len(y_test)) 
print(y_pred.shape,y_test.shape)

from sklearn import metrics  
import numpy as np  
sum_mean=0  
for i in range(len(y_pred)):  
    sum_mean+=(y_pred[i]-y_test.values[i])**2  
sum_erro=np.sqrt(sum_mean/50)  


print("RMSE by hand:",sum_erro) 
```

```python
from sklearn.metrics import mean_squared_error
print("RMSE:", mean_squared_error(y_pred, y_test)**0.5)

from sklearn.metrics import r2_score
print('r2 score:',r2_score(y_pred, y_test))
```

```python
import matplotlib.pyplot as plt  
plt.figure()  
plt.plot(range(len(y_pred)),y_pred,'b',label="predict")  
plt.plot(range(len(y_pred)),y_test,'r',label="test")  
plt.legend(loc="upper right")
plt.xlabel("VND/HKD actual rate")  
plt.ylabel('USD pred rate')  
plt.show()  
```
